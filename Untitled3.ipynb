# 10.  What is the relationship between selling_price and km_driven for first-owner bikes ?

```python
import matplotlib.pyplot as plt
import seaborn as sns


first_owner_bikes = df[df['owner'] == '1st owner']

# Scatter plot with regression line
plt.figure(figsize=(8, 5))
sns.regplot(
    x='km_driven',
    y='selling_price',
    data=first_owner_bikes,
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'red'}
)
plt.title('Relationship between Selling Price and KM Driven for First-Owner Bikes')
plt.xlabel('KM Driven')
plt.ylabel('Selling Price (INR)')
plt.show()

# Correlation coefficient
correlation = first_owner_bikes[['km_driven', 'selling_price']].corr().iloc[0, 1]
print(f"Correlation between KM Driven and Selling Price for first-owner bikes: {correlation:.2f}")
```

# 11.  Identify and remove outliers in the km_driven column using the IQR method ?

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Determine the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and remove outliers
filtered_df = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Print the results
print(f"Original dataset size: {df.shape[0]} rows")
print(f"Filtered dataset size: {filtered_df.shape[0]} rows")
print(f"Number of outliers removed: {df.shape[0] - filtered_df.shape[0]} rows")
```

# 13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing year) ?


```python
import pandas as pd

# Add a new column for bike's age
current_year =
df['bike_age'] = current_year - df['year']

# Calculate the average selling price for each bike age
avg_price_by_age = df.groupby('bike_age')['selling_price'].mean().reset_index()

# Calculate depreciation per year
avg_price_by_age['depreciation_per_year'] = avg_price_by_age['selling_price'].diff().fillna(0)

# Display the result
print(avg_price_by_age[['bike_age', 'selling_price', 'depreciation_per_year']])
```

# 14.  Which bike names are priced significantly above the average price for their manufacturing year ?

```python
# Calculate the average selling price for each manufacturing year
avg_price_by_year = df.groupby('year')['selling_price'].mean().reset_index()
avg_price_by_year.rename(columns={'selling_price': 'avg_year_price'}, inplace=True)

# Merge the average price data with the original dataset to compare each bike's price with the average price of its year
df = pd.merge(df, avg_price_by_year, on='year', how='left')

# Define the threshold for "significantly above average" (e.g., 20% higher than the average)
threshold = 0.2  # 20% above the average

# Identify bikes that are priced significantly above the average for their manufacturing year
significantly_above_avg = df[df['selling_price'] > (1 + threshold) * df['avg_year_price']]

# List the bike names that meet the condition
significantly_above_avg_bikes = significantly_above_avg['name'].unique()

# Display the result
print("Bikes priced significantly above the average price for their manufacturing year:")
for bike in significantly_above_avg_bikes:
    print(bike)
```

# 1. What is the average selling price of cars for each dealer, and how does it compare across different dealers ?

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Group by Dealer_Name and calculate the average selling price
average_price_per_dealer = df.groupby('Dealer_Name')['Price ($)'].mean().reset_index()

# Sort by price to see which dealers have the highest/lowest average prices
average_price_per_dealer = average_price_per_dealer.sort_values('Price ($)', ascending=False)

# Display the result
print(average_price_per_dealer)

# Visualize the comparison with a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=average_price_per_dealer, x='Dealer_Name', y='Price ($)', palette='viridis')
plt.title('Average Selling Price of Cars for Each Dealer')
plt.xticks(rotation=70)  # Rotate dealer names for better readability
plt.ylabel('Average Price ($)')
plt.xlabel('Dealer Name')
plt.show()
```

# 12.  Is there any seasonal trend in car sales based on the date of sale ?

```python
# Ensure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year from the date
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Group data by month and calculate total sales
monthly_sales = df.groupby('Month').size().reset_index(name='Number of Sales')

# Plot the number of cars sold per month
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Number of Sales', marker='o', color='teal')
plt.title('Seasonal Trend in Car Sales (Monthly)')
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Group data by year and month to analyze trends over time
year_month_sales = df.groupby(['Year', 'Month']).size().reset_index(name='Number of Sales')

# Heatmap of car sales across years and months
sales_pivot = year_month_sales.pivot(index='Year', columns='Month', values='Number of Sales')

plt.figure(figsize=(12, 8))
sns.heatmap(sales_pivot, annot=True, fmt='.0f', cmap='YlGnBu', cbar_kws={'label': 'Number of Cars Sold'})
plt.title('Car Sales Heatmap by Year and Month')
plt.ylabel('Year')
plt.xlabel('Month')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

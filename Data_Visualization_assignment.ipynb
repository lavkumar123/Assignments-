# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generating a synthetic dataset
np.random.seed(42)  # For reproducibility
n = 100  # Number of data points

# Generating two variables with some correlation
x = np.random.normal(50, 10, n)  # Independent variable
y = 2.5 * x + np.random.normal(0, 15, n)  # Dependent variable with some noise

# Creating a DataFrame
data = pd.DataFrame({'Variable X': x, 'Variable Y': y})

# Creating the scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Variable X', y='Variable Y', data=data, color='b', edgecolor='w', s=70)
plt.title("Scatter Plot of Relationship Between Variable X and Variable Y")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generating a synthetic dataset with correlated features
np.random.seed(42)
n = 100  # Number of samples

# Generating correlated features
x = np.random.normal(0, 1, n)  # Feature 1
y = 2 * x + np.random.normal(0, 1, n)  # Feature 2 (correlated with x)
z = -x + np.random.normal(0, 1, n)  # Feature 3 (inverse correlation with x)
w = 0.5 * y + np.random.normal(0, 1, n)  # Feature 4 (correlated with y)

# Creating a DataFrame
df = pd.DataFrame({'Feature 1': x, 'Feature 2': y, 'Feature 3': z, 'Feature 4': w})

# Calculating the correlation matrix
corr_matrix = df.corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.


```python
import numpy as np
import pandas as pd
import plotly.express as px

# Set the random seed for reproducibility
np.random.seed(20)

# Create the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Generate the heatmap
fig = px.imshow(sales_pivot,
                title="Sales Variation Across Months and Days",
                labels={'color': 'Average Sales'},
                x=sales_pivot.columns,
                y=sales_pivot.index,
                color_continuous_scale="Viridis")

# Show the plot
fig.show()
```

# 4. Using the given x and y data, generate a 3D surface plot to visualize the function.

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values based on the function
z = np.sin(np.sqrt(x**2 + y**2))

# Create the surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale="Viridis")])

# Update layout for better visualization
fig.update_layout(title="3D Surface Plot of z = sin(√(x² + y²))",
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'
                  ))

# Show the plot
fig.show()
```

# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.


```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Enable the output to show in the notebook
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(10, 30, 100)  # Random sizes for markers
colors = np.random.rand(100) * 100       # Random values for color mapping

# Create a data source for Bokeh
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot
p = figure(title="Random Scatter Plot with Varying Sizes and Colors",x_axis_label="x", y_axis_label="y")

# Set up color mapping based on the 'colors' column
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Add scatter renderer
p.scatter('x', 'y', size='sizes', color=color_mapper, source=source, legend_label="Data Points", alpha=0.6)

#
```

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable Bokeh to display in the notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a data source for Bokeh
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label="Fruits", y_axis_label="Counts")

# Add vertical bars to the figure
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="skyblue")

# Show the plot
show(p)
```

# 4. Create a Bokeh histogram to visualize the distribution of the given data.



```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable Bokeh output in the notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Create histogram data (frequencies and bin edges)
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:], bottom=np.zeros_like(hist)))

# Create a figure for the histogram
p = figure(title="Histogram of Data", x_axis_label="Value", y_axis_label="Frequency")

# Add a quadrilateral glyph to represent the bars
p.quad(top='top', bottom='bottom', left='left', right='right', source=source,
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap

# Enable Bokeh output in the notebook
output_notebook()

# Generate dataset
data_heatmap = np.random.rand(10, 10)  # Random values for heatmap
x = np.linspace(0, 1, 10)  # x-axis values
y = np.linspace(0, 1, 10)  # y-axis values
xx, yy = np.meshgrid(x, y)  # Create meshgrid

# Flatten the meshgrid and data to create a ColumnDataSource
x_vals = xx.flatten()
y_vals = yy.flatten()
colors = data_heatmap.flatten()

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(data=dict(x=x_vals, y=y_vals, colors=colors))

# Create a figure
p = figure(title="Heatmap with Levels", x_axis_label="X", y_axis_label="Y")

# Use linear_cmap to map data values to a color palette
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Add rect glyphs to represent each cell in the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       color=color_mapper, line_color=None)

# Add color bar to represent levels of the heatmap
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
